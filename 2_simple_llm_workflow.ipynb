```python
from langgraph.graph import StateGraph, START,END
from langchain_openai import ChatOpenAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
import os
```

```python
model= ChatOpenAI(api_key=os.getenv("LANGCHAIN_API_KEY"))
```

```python
class LLMState(TypedDict):
    question:str
    ans:str
```

```python
def llm_qa(state:LLMState)->LLMState:
    question=state["question"]
    prompt= f'Answer the following question in detail: {question}'
    response= model.invoke(prompt)
    state["ans"]= response.content
    return state
```

```python
graph= StateGraph(LLMState)

graph.add_node("llm_qa", llm_qa)

graph.add_edge(START, "llm_qa")
graph.add_edge("llm_qa", END)

workflow=graph.compile()
```

```python
initial_state= LLMState(question="What is LangGraph?")
```

```python
final_state=workflow.invoke(initial_state)
```

```python
print(final_state["ans"])
```
